#### The variable Cabin has far too many NAs for extrapolation. So, the column is removed. For Age, the missing values will be populated with median Age of the entire column. For Embarked, the rows with NA will be removed since it is only 2 out of 890.

```python
# Removing the column Cabin
df.drop('Cabin', axis=1, inplace=True)
```

```python
# Filling NAs in column Age with median age
median_age = statistics.median(df['Age'].dropna())
print(median_age)
df['Age'] = df['Age'].fillna(median_age)
```

```python
# Removing final 2 NA rows from Column Embarked
df = df.dropna()
```
